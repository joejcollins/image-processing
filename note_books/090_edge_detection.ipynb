# Edge Detection

The edge is not actually that sharp

```python
import sys

import imageio.v3 as iio
import matplotlib.pyplot as plt
import skimage.feature

# read command-line arguments
filename = "../data/raw/training/shapes-01.jpg"
sigma = 1
low_threshold = 0.05
high_threshold = 0.07

image = iio.imread(uri=filename, mode="L")
plt.imshow(image)
```

```python
edges = skimage.feature.canny(
    image=image,
    sigma=sigma,
    low_threshold=low_threshold,
    high_threshold=high_threshold,
)
# display edges
skimage.io.imshow(edges)
# display edges
fig, ax = plt.subplots()
ax.imshow(edges)
```
