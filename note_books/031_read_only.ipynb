# Read Only Arrays

Sometimes the array that comes from the `iio.imread` is read only.
Like if the storage is a memory mapped file and
you don't want modifications to be made.
I guess you could set `.flags.writeable = True`
but you would be making a mess.
Better to use `np.array()` to make the image into an array.

But why not use `.copy()`?

```python
import imageio.v3 as iio
import numpy
from image_processing import file_finder_service
from matplotlib import pyplot

maize_image_path = file_finder_service.find_training_image("maize-root-cluster.jpg")

maize_roots = iio.imread(uri=maize_image_path)
maize_roots.flags.writeable = True
writeable_maize_roots = numpy.array(maize_roots)

# display original image
fig, ax = pyplot.subplots(figsize=(15,15))
ax.imshow(writeable_maize_roots)
```
