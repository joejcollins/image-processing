```python
%matplotlib widget

import imageio.v3 as iio
import ipympl
import matplotlib.pyplot as plt
import numpy as np
import skimage as ski
from aqua_marina import file_finder_service

plant_seedling_path = file_finder_service.find_training_image('plant-seedling.jpg')
plant_seedling  = iio.imread(plant_seedling_path)
# display the image
fig, ax = plt.subplots()
ax.imshow(plant_seedling)
```

```python
# tuple to select colors of each channel line
colors = ("red", "green", "blue")

fig, ax = plt.subplots()
ax.set_xlim([0, 256])
for channel_id, color in enumerate(colors):
    histogram, bin_edges = np.histogram(
        plant_seedling[:, :, channel_id], bins=256, range=(0, 256)
    )
    ax.plot(bin_edges[:-1], histogram, color=color)

ax.set_title("Color Histogram")
ax.set_xlabel("Color value")
ax.set_ylabel("Pixel count")
```
