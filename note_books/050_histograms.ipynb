# Histograms


```python
%matplotlib widget

import imageio.v3 as iio
import ipympl
import matplotlib.pyplot as plt
import numpy as np
import skimage as ski
from image_processing import file_finder_service, image_display

plant_seedling_path = file_finder_service.find_training_image('plant-seedling.jpg')
plant_seedling  = iio.imread(plant_seedling_path, mode='L')

plant_seedling = ski.util.img_as_float(plant_seedling)
# display the image
fig, ax = plt.subplots(figsize=(10, 10))
ax.imshow(plant_seedling, cmap="gray")
histogram, bin_edges = np.histogram(plant_seedling, bins=256, range=(0, 1))
```

```python
# configure and draw the histogram figure
fig, ax = plt.subplots()
ax.set_title("Grayscale Histogram")
ax.set_xlabel("grayscale value")
ax.set_ylabel("pixel count")
ax.set_xlim([0.0, 1.0])

ax.plot(bin_edges[:-1], histogram)
```

```python
rr, cc = ski.draw.rectangle(start=(200, 445), end=(325, 485))
mask = np.zeros(shape=plant_seedling.shape, dtype=bool)
mask[rr, cc] = True
fig, ax = plt.subplots()
ax.imshow(mask, cmap="gray")
```

```python
histogram, bin_edges = np.histogram(plant_seedling[mask], bins=256, range=(0.0, 1.0))
fig, ax = plt.subplots()

ax.set_title("Grayscale Histogram")
ax.set_xlabel("grayscale value")
ax.set_ylabel("pixel count")
ax.set_xlim([0.0, 1.0])
ax.plot(bin_edges[0:-1], histogram)
```
