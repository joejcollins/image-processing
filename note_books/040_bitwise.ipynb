# Drawing and bit wise images

Focusing on bits of the images.


```python
%matplotlib widget

import imageio.v3 as iio
import ipympl
import matplotlib.pyplot as plt
import numpy as np
import skimage
import skimage.draw
from image_processing import file_finder_service, image_display

```

```python
seedlings_path = file_finder_service.find_training_image("maize-seedlings.tif")
seedlings = image = iio.imread(uri=seedlings_path)

fig, ax = plt.subplots()
ax.imshow(seedlings)

mask = np.ones(shape=seedlings.shape[:2], dtype=bool)
```

```python
# Draw filled rectangle on the mask image
rr, cc = skimage.draw.rectangle(start=(357, 44), end=(740, 720))
mask[rr, cc] = False

fig, ax = plt.subplots()
ax.imshow(mask, cmap="gray")
```

```python
canvas = np.zeros(shape=(600,800,3), dtype=np.uint8)

rr, cc = skimage.draw.disk(
    center=(200, 300), radius=100, shape=canvas.shape[:2]
)
canvas[rr, cc] = (0, 0, 255)

fig, ax = plt.subplots()
ax.imshow(canvas)
```

```python
seedlings[mask] = 0

fig, ax = plt.subplots()
ax.imshow(seedlings)
```
