```python
import imageio.v3 as iio
import ipympl
import matplotlib.pyplot as plt
import skimage as ski

%matplotlib widget
from aqua_marina import file_finder_service

image_path = file_finder_service.find_training_image("gaussian-original.png")
image = iio.imread(image_path)

# display the image
fig, ax = plt.subplots()
ax.imshow(image)
```

```python
sigmay = 20.0
sigmax = 0

# apply Gaussian blur, creating a new image
blurred = ski.filters.gaussian(
    image, sigma=(sigmay, sigmax), truncate=3.5, channel_axis=-1)

# display the image
fig, ax = plt.subplots()
ax.imshow(blurred)
```
