# Grayscale

The function `ski.color.rgb2gray()` adds up the three colour channels
in a way that matches human colour perception,

```python
import imageio.v3 as iio
from aqua_marina import file_finder_service
from matplotlib import pyplot

chair_path = file_finder_service.find_training_image("chair.jpg")
chair_picture = iio.imread(uri=chair_path)

# display original image
fig, ax = pyplot.subplots()
ax.imshow(chair_picture)
```

Use a cmap to make it actually 

```python
import skimage

chair_in_grayscale = skimage.color.rgb2gray(chair_picture)

fig, ax = pyplot.subplots()
ax.imshow(chair_in_grayscale, cmap="gray")
```
